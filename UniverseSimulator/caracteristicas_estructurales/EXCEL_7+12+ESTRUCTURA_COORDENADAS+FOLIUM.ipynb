```python
import pandas as pd
```

```python
df_7 = pd.read_csv("df_7.csv").iloc[:, 1:]
df_7.head()
```

```python
df_7["AFECBLAN"] = [1 if x == "SI" else 0 for x in df_7["AFECBLAN"]]
df_7["AFECGRIS"] = [1 if x == "SI" else 0 for x in df_7["AFECGRIS"]]
```

```python
df_7.head()
```

```python
df_7 = df_7.drop(["Shape_Length", "Shape_Area", "CUEN_ZB"], axis = 1)
df_7.head()
```

```python
df_12 = pd.read_csv("df_12.csv").iloc[:, 1:]
df_12.head()
```

```python
df_12 = df_12.rename({'codMuni':'CODMUN'}, axis=1)
df_12
```

```python
df_estructura = pd.read_csv("df_estructura.csv").iloc[:, 1:]
df_estructura.head()
```

```python
df_estructura = df_estructura.rename({'CODMUNI':'CODMUN'}, axis=1)
df_estructura
```

```python
df_coordenadas = pd.read_csv("df_coordenadas.csv").iloc[:, 1:]
df_coordenadas
```

```python
from functools import reduce
df_list = [df_7, df_12, df_estructura, df_coordenadas]
df_merged = reduce(lambda left,right: pd.merge(left,right,on=['CODMUN'],
                   how='outer'), df_list)
```

```python
df_merged
```

```python
# Save local
df_merged.to_csv('df_estructura_completo.csv') 
```
